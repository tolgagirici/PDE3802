```python
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib inline
```

```python
numbers = np.random.randint(1,100, (1,10))
print(numbers)
maximum = 0
minimum = np.inf
for i in numbers[0,:]:
    #print(i)
    if i>maximum:
        maximum = i
    if i<minimum:
        minimum = i
print(maximum, minimum)

print(np.max(numbers), np.min(numbers))
```

```python
t=np.linspace(-np.pi, +np.pi, 100)
f = np.sin(t)
f2=np.cos(t)
f3= np.tan(t)
f4= np.atan(t)

plt.subplot(2,2,1)
plt.plot(t,f,"b-.")
plt.grid("on")
plt.ylabel("cos")

plt.subplot(2,2,2)
plt.plot(t,f2, "k--")
plt.xlabel("t")
plt.ylabel("sin")
plt.grid("on")

plt.subplot(2,2,3)
plt.plot(t,f3, "k--")
plt.xlabel("t")
plt.ylabel("tan")
plt.grid("on")

plt.subplot(2,2,4)
plt.plot(t,f4, "k--")
plt.xlabel("t")
plt.ylabel("arctan")
plt.grid("on")
plt.show()
```
